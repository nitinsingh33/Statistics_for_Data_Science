# PMF & CDF Plotting

```python
import numpy as np
import matplotlib.pyplot as plt
from math import comb, factorial

def binomial_dis(n,p):
    x=np.arange(0,n+1)
    pmf=[comb(n,k)*(p**k)*((1-p)**(n-k)) for k in x]
    cdf=np.cumsum(pmf)
    return x, pmf, cdf
```

```python
def plot_distribution(x,pmf,cdf):
    plt.figure(figsize=(12,6))

    # Ploting of PMF
    plt.subplot(1,2,1)
    plt.stem(x,pmf)
    plt.title('PMF')
    plt.xlabel('Number of successes')
    plt.ylabel('P(X=k)')

    # Ploting of CDF
    plt.subplot(1,2,2)
    plt.step(x,cdf)
    plt.title('cdf')
    plt.xlabel('Number of successes')
    plt.ylabel('P(X<=k)')
```

```python
n_b, p_b=10, 0.5
x_binomial, pmf_binomial, cdf_binomial = binomial_dis(n_b, p_b)
plot_distribution(x_binomial,pmf_binomial,cdf_binomial)
```
